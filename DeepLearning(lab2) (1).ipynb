```python
# perceptron from scratch
# Step 1: Input data
X = [0.75, 0.65, 0.80]      # Input features
w = [0.4, 0.3, 0.5]        # Initial weights
b = -0.2                  # Initial bias
Y = 1                     # Target output
n = 0.01                  # Learning rate

print("Initial Weights:", w)
print("Initial Bias:", b)


# Step 2: Compute net input
z = 0
for i in range(len(X)):
    z += w[i] * X[i]
z += b

print("\nNet Input (z):", z)


# Step 3: Activation function
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted Output:", y_pred)


# Step 4: Calculate error
loss = Y - y_pred
print("Loss:", loss)


# Step 5: Check prediction
if loss == 0:
    print("Prediction matches target, No Update required")
else:
    print("Prediction is wrong")

    # Step 6: Update weights
    for i in range(len(w)):
        w[i] = w[i] + n * loss * X[i]

    # Step 7: Update bias
    b = b + n * loss

    # Recalculate net input after learning
    z_new = 0
    for i in range(len(X)):
        z_new += w[i] * X[i]
    z_new += b

    print("New net input (z):", z_new)

    if z_new >= 0:
        final_output = 1
    else:
        final_output = 0

    print("Final Prediction:", final_output)


# Step 8: Print updated results
print("\nUpdated Weights:", w)
print("Updated Bias:", b)
```
